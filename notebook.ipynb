```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer

# 1. Load Data
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target  # 0 = Malignant, 1 = Benign

# 2. EDA: Target Distribution (Required Chart 1)
sns.countplot(x=df['target'])
plt.title("Class Distribution (0=Malignant, 1=Benign)")
plt.show()

# 3. EDA: Feature Correlation (Required Chart 2)
plt.figure(figsize=(10,6))
sns.scatterplot(x='mean radius', y='mean texture', hue='target', data=df, alpha=0.6)
plt.title("Mean Radius vs Mean Texture")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 1. Split Data
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)

# 2. Model A: Logistic Regression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, lr.predict(X_test)):.4f}")

# 3. Model B: Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf.predict(X_test)):.4f}")
```
